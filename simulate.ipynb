```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

root = Path.cwd()
figures = root / "figures"
```

```python
df = pd.read_csv(root / "1.dat")
ax, ay, bx, by = df.iloc[:, 0], df.iloc[:, 1], df.iloc[:, 2], df.iloc[:, 3]

fig, axes = plt.subplots()
axes.plot(ax, ay, label="Galaxy A")
axes.plot(bx, by, label="Galaxy B")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig(figures / "1.png")
plt.show()
```

```python
df = pd.read_csv(root / "2.dat")
ax, ay, az, bx, by, bz = df.iloc[:, 0], df.iloc[:, 1], df.iloc[:, 2], df.iloc[:, 3], df.iloc[:, 4], df.iloc[:, 5]

plt.plot(ax, ay, label = 'Mass 1')
plt.plot(bx, by, label = 'Mass 2')
plt.title('Trajectories of Two Body System')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig(figures / "2.png")
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
axes = fig.add_subplot(111, projection='3d')
axes.plot(ax, ay, az, label = 'Mass 1')
axes.plot(bx, by, bz, label = 'Mass 2')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_xlabel('z')
axes.legend()
plt.title('3D Trajectory')
plt.savefig(figures / "3.png")
plt.show()
```
